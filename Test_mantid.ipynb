```python
def ExtractRawData(file_path):
    raw_data = ms.Load(file_path)
    raw_north_data = ms.CropWorkspace(InputWorkspace = raw_data, StartWorkspaceIndex=0,EndWorkspaceIndex = 1199)
    raw_north_spacing = ms.ConvertUnits(InputWorkspace = raw_north_data, Target = "dSpacing", AlignBins=True)
    raw_north_summed = ms.SumSpectra(raw_north_spacing)
    North_neutron_X_data = raw_north_summed.dataX(0)
    North_neutron_Y_data = raw_north_summed.dataY(0)
    raw_south_data = ms.CropWorkspace(InputWorkspace = raw_data, StartWorkspaceIndex=1200, EndWorkspaceIndex = 2399)
    raw_south_spacing = ms.ConvertUnits(InputWorkspace = raw_south_data, Target = "dSpacing", AlignBins=True)
    raw_south_summed = ms.SumSpectra(raw_south_spacing)
    South_neutron_X_data = raw_south_summed.dataX(0)
    South_neutron_Y_data = raw_south_summed.dataY(0)
    start_time = str(raw_data.run().startTime())
    end_time = str(raw_data.run().endTime())
    Title = raw_data.getTitle()
    return(North_neutron_X_data,North_neutron_Y_data,South_neutron_X_data,South_neutron_Y_data,start_time,end_time,Title)
```

```python
%%timeit
directory = r"E:\cycle_04_1"
output = r"E:\temp_raw_data"

for filename in os.listdir(directory):
    file_path = directory+"\\"+filename
    if "Copy" in filename:
        pass
    elif "Shortcut" in filename:
        pass
    elif filename[-3:].lower() == "raw":
        print(filename)
        extracted_raw_data = ExtractRawData(file_path)
        print(filename[3:8])
        output_file = open(output + "\\" + str(filename[3:8]) + ".txt", "w+")
        for data in extracted_raw_data:
            output_file.write("{}\n".format(data))
            output_file.write("kamehameha")#SOMTHING THAT WILL NEVER COME UP IN THE DATA
        output_file.close()
        
```
