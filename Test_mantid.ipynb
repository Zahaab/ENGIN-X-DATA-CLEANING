```python
import os 
import math
import mantid.simpleapi as ms
from multiprocessing.dummy import Pool as ThreadPool 
```

```python
def ExtractRawData(file_path):
    raw_data = ms.Load(file_path)
    raw_north_data = ms.CropWorkspace(InputWorkspace = raw_data, StartWorkspaceIndex=0,EndWorkspaceIndex = 1199)
    raw_north_spacing = ms.ConvertUnits(InputWorkspace = raw_north_data, Target = "dSpacing", AlignBins=True)
    raw_north_summed = ms.SumSpectra(raw_north_spacing)
    North_neutron_X_data = raw_north_summed.dataX(0)
    North_neutron_Y_data = raw_north_summed.dataY(0)
    raw_south_data = ms.CropWorkspace(InputWorkspace = raw_data, StartWorkspaceIndex=1200, EndWorkspaceIndex = 2399)
    raw_south_spacing = ms.ConvertUnits(InputWorkspace = raw_south_data, Target = "dSpacing", AlignBins=True)
    raw_south_summed = ms.SumSpectra(raw_south_spacing)
    South_neutron_X_data = raw_south_summed.dataX(0)
    South_neutron_Y_data = raw_south_summed.dataY(0)
    start_time = str(raw_data.run().startTime())
    end_time = str(raw_data.run().endTime())
    Title = raw_data.getTitle()
    return(North_neutron_X_data,North_neutron_Y_data,South_neutron_X_data,South_neutron_Y_data,start_time,end_time,Title)
```

```python
def CreateFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
    return directory

def FindDel(S):
    """This function takes a the first string in a list,(S) that starts with ENG, it finds the nomancature 
    the file uses for the begining of each file, e.g ENG44007 or ENGINX00305258, it shall also work if more numbers
    are added or the name changes so long as ENG followed by anything other than "." or "_" or " " is still 
    the start of each file"""
    for i in S:
        if i[0:3] == 'ENG':
            n = 1
            for j in i:
                if j == '.':
                    break
                elif j == ' ':
                    break
                elif j == '_':
                    break
                else: 
                    n += 1 
            return n
        else:
            continue

def RunIndex(s,n): #ENGINX00305258
    """s is a string in the for m of a filename from the enginx data and returns it's run number, n is the FindDel"""
    last_int_index = n-1
    first_int_index = 0
    for j in range(last_int_index):
        try:
            int(s[j])
            return s[first_int_index:last_int_index]
        except:
            first_int_index += 1
```

```python
%%time
directory = r"C:\Users\zahaab\test_case4"
output = CreateFolder(directory+"\\temp_raw_data")

for foldername in os.listdir(directory):
    
    CreateFolder(output+"\\"+foldername)
    folder_path = directory+"\\"+foldername
    end_path = output+"\\"+foldername
    raw_files = [value for value in os.listdir(folder_path) if value[-3:] in ["RAW", "raw", "Raw"]]
    n = FindDel(os.listdir(folder_path))
    
    for filename in raw_files:
        file_path = folder_path+"\\"+filename
        
        if "Copy" in filename:
            continue
                
        elif filename == os.listdir(directory)[-1]:
            continue
        
        elif filename[n] in ("S", "s", "N", "n"):# Stop temporary files from entering (e.g .s01 or .n001)
                try:
                    int(filename[n+1]) # all temp files are 's' or 'n' followed by a int
                    continue
                except ValueError: 
                    pass
                
        if filename[-3:].lower() == "raw":
            print(filename)
            extracted_raw_data = ExtractRawData(file_path)
            run_index = RunIndex(filename, n)
            print(run_index)
            output_file = open(end_path + "\\" + str(run_index) + ".txt", "w+")
            for data in extracted_raw_data:
                output_file.write("{}\n".format(data))
                output_file.write("kamehameha")#SOMTHING THAT WILL NEVER COME UP IN THE DATA
            output_file.close()
```
